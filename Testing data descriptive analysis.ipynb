```python
import pandas as pd
df = pd.read_csv('water_potability.csv')
df.head(5)
```

```python
from matplotlib import pyplot
pyplot.plot(df['Conductivity'], df['Potability'])
pyplot.show()
```

```python
for column in df:
    pyplot.hist(df[column], bins=100)
    pyplot.xlabel(column)
    pyplot.show()
```

```python
for column in df:
    pyplot.boxplot(df[column].dropna())
    pyplot.xlabel(column)
    pyplot.show()
```

```python
for column in df:
    pyplot.scatter(df["Potability"], df[column])
    pyplot.xlabel(column)
    pyplot.show()
```

```python
pd.plotting.scatter_matrix(df, alpha=0.2)
```
